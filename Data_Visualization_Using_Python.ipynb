**Importing necessary modules**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
#iris dataset
iris=snr.load_dataset("iris")
```

```python
iris.head()
```

```python
iris.value_counts('species') #setosa #virginica #versicolor
```

```python
iris.tail()
```

### **Line graph**

```python
a=iris.plot(figsize=(10,5),title='Line graph on iris data')
a.set_xlabel('x_axis')#rows
a.set_ylabel('y_axis')#columns
```

```python
b=iris['petal_length'].plot(figsize=(10,5),title='Line graph on iris data')
```

```python
b=iris['sepal_length'].plot(figsize=(10,5),title='Line graph on iris data')
```

```python
b=iris['sepal_width'].plot(figsize=(10,5),title='Line graph on iris data')
```

```python
b=iris['petal_width'].plot(figsize=(8,2),title='Line graph on iris data')
```
